```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\subhu\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get('https://www.flipkart.com/')
```

```python
product=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
product.send_keys('sneakers')
```

```python
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[2]/div[1]/div[2]/div[2]/form/div/button')
search.click()
```

```python
Brand=[]
Product_Description=[]
Price=[]
```

```python
b_tag=driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
for i in b_tag[0:36]:
    brand=i.text
    Brand.append(brand)
    
pro_tag=driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
for i in pro_tag[0:36]:
    product=i.text
    Product_Description.append(product)
pri_tag=driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
for i in pri_tag[0:36]:
    prices=i.text
    Price.append(prices)
```

```python
print(len(Price),len(Product_Description),len(Brand))
```

```python
df1=pd.DataFrame({'Brand':Brand,'Price':Price,'Product_Description':Product_Description})
```
