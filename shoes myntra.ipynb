```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
driver=webdriver.Chrome(r"C:\Users\subhu\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get('http://www.myntra.com/shoes')
```

```python
filter1=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label')
filter1.click()
```

```python
filter2=driver.find_element(By.XPATH,'/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label')
filter2.click()
```

```python
Brand=[]
description=[]
Price=[]
```

```python
b_tag=driver.find_elements(By.XPATH,'//h3[@class="product-brand"]')
for i in b_tag:
    brand=i.text
    Brand.append(brand)
    
des_tag=driver.find_elements(By.XPATH,'//h4[@class="product-product"]')
for i in des_tag:
    product=i.text
    description.append(product)
pri_tag=driver.find_elements(By.XPATH,'//div[@class="product-price"]')
for i in pri_tag:
    prices=i.text
    Price.append(prices)
```

```python
print(len(description),len(Brand),len(Price))
```

```python
df=pd.DataFrame({'Brand':Brand,'description':description,'Price':Price})
```
